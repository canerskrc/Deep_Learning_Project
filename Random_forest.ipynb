```python
# Viraj açısı, yol durumu, hız, görünürlük, trafik yoğunluğu
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
np.random.seed(42)
data_size = 200 # veriseti boyutu

```

```python

data = {
    "virajacisi": np.random.randint(10,180, size=data_size),
    "yol_durumu":np.random.choice(["Kuru","islak","buzlu"],size=data_size),
    "hiz":np.random.randint(30,150, size=data_size),
    "gorunurluk":np.random.choice(["gece","gunduz"],size=data_size),
    "trafik_yogunlugu":np.random.randint(1,5,size=data_size)
}
df = pd.DataFrame(data)
# hedef değişken : KAza olma durumu ( 1= kaza, 0 = kaza yok)

df["Kaza_Durumu"] = [
    1 if (row["virajacisi"]> 100 and row["hiz"] > 80) or row["yol_durumu"]=="buzlu" else 0
    for index, row in df.iterrows()
]
df.head()
```

```python
df["yol_durumu"]= df["yol_durumu"].map({"Kuru":0, "islak":1, "buzlu":2})
df["gorunurluk"]=df["gorunurluk"].map({"gece":0,"gunduz":1})
df.head()
```

```python
X = df.drop("Kaza_Durumu", axis=1)
y = df["Kaza_Durumu"]
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=42)

rf_model = RandomForestClassifier(n_estimators = 100, random_state=42) # n_estimators = kullanılacak karar ağacı sayısı
rf_model.fit(X_train,y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
report = classification_report(y_test, y_pred)

print(f"Doğruluk: {accuracy}\n")
print("Sınıflandırma Raporu:")
print(report)
```

Support sütunu her sınıfa ait gerçek verileri gösterir.

Precision = Kesinlik Yapılan tahminlerin doğruluğunu ölçer.

TP / (TP + FP )

Recall = Bir sınıfın gerçek verilerini ne kadar iyi yakalayabildiğini ölçer.

TP / (TP + FN )

F1 Score = Precision ve Recall değerlerinin harmonik ortalaması.

 2 x ( Pre x ReCll ) / ( Pre + ReCll )
